```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
adni_data = pd.read_csv('ADNIMERGE_20Aug2024.csv')

# 1. Check for missing values in each column
missing_values = adni_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python

# 2. Summary statistics for key numeric columns
summary_stats = adni_data.describe()
print("Summary statistics:\n", summary_stats)

```

```python
# 3. Visualize the distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(adni_data['AGE'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python

# 4. Visualize the distribution of Diagnostic Categories (DX)
plt.figure(figsize=(8, 6))
sns.countplot(x='DX', data=adni_data)
plt.title('Diagnostic Category Distribution')
plt.xlabel('Diagnostic Categories')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Visualize the correlation matrix (if there are numerical variables)
plt.figure(figsize=(12, 8))
sns.heatmap(adni_data.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# 5. Visualize the correlation matrix (if there are numerical variables)
plt.figure(figsize=(12, 8))
# Select only numerical columns before calculating correlation
sns.heatmap(adni_data.select_dtypes(include=['number']).corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
