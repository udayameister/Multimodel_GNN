```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns

# Number of regions (nodes) - AAL regions for brain connectivity
num_regions = 116
embedding_dim = 64  # Example embedding dimension

# Generate random sample embeddings for different models (replace with real embeddings)
# Multimodal GAT
gat_multimodal_embeddings = np.random.randn(num_regions, embedding_dim)
# Unimodal GAT
gat_unimodal_embeddings = np.random.randn(num_regions, embedding_dim)
# GCN Embeddings
gcn_embeddings = np.random.randn(num_regions, embedding_dim)
# Node2Vec Embeddings
node2vec_embeddings = np.random.randn(num_regions, embedding_dim)

# Generate random labels (4 classes: CN, EMCI, LMCI, AD) for the example
# In practice, you should use your actual class labels
labels = np.random.randint(0, 4, num_regions)

# Mapping from label integers to actual class names for better visualization
class_names = {0: 'CN', 1: 'EMCI', 2: 'LMCI', 3: 'AD'}

# List of model embeddings
embeddings_list = [
    ('GAT Multimodal', gat_multimodal_embeddings),
    ('GAT Unimodal', gat_unimodal_embeddings),
    ('GCN', gcn_embeddings),
    ('Node2Vec', node2vec_embeddings)
]

# Convert label integers to class names
labels_class = np.vectorize(class_names.get)(labels)

# Function to apply t-SNE and plot
def plot_tsne(embeddings, labels, title):
    # Apply t-SNE
    tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42)
    tsne_results = tsne.fit_transform(embeddings)

    # Create scatter plot
    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        x=tsne_results[:, 0], y=tsne_results[:, 1],
        hue=labels, palette=sns.color_palette("hsv", len(np.unique(labels))),
        legend="full", s=60, alpha=0.8
    )
    plt.title(f't-SNE Visualization - {title}')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Loop through each model and visualize embeddings using t-SNE
for model_name, embeddings in embeddings_list:
    plot_tsne(embeddings, labels_class, model_name)
```
